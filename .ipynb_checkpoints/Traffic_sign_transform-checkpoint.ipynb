```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline

# read in and display the original image
img = mpimg.imread('./stopsign.jpg')

plt.imshow(img)
```

```python

plt.imshow(img)

# points to define a rectangle plane on the image
# display points onto the image
plt.plot(940, 350, '.', color='r') # top right
plt.plot(940, 730, '.', color='r') # bottom right
plt.plot(180, 750, '.', color='r') # bottom left
plt.plot(200, 420, '.', color='r') # top left
```

```python
# funstion takes in an image and returns a warped image
# perspective transform function
def warp(img):
    
    
    
    # define image size x, y
    img_size = (img.shape[1], img.shape[0])
    
    # define calibration box in source (original) and destination (corrected) coordinates
    
    # four coordinates from source image sign corners
    src = np.float32(
        [[940,350],
         [940,730],
         [180,750],
         [200,420]])
    
    # for points to define the new rectangle corrected image position
    dst = np.float32(
        [[940,400],
         [940,750],
         [200,750],
         [200,400]])
    
    # computer perspective transform return a matrix of the transform
    M = cv2.getPerspectiveTransform(src, dst)
    
    # example - can get the inverse from corrected to original by swapping soruce with destination:
    #Minv = cv2.getPerspectiveTransform(dst, src)
    
    # apply transform M to original image to obtain warped image
    # returns a warped image same size as original image using transform M
    # flags tells it how to interpolate points - fill in missing points
    warped = cv2.warpPerspective(img, M, img_size, flags=cv2.INTER_LINEAR)
    
    return warped
    

warped_image = warp(img)

```

```python
%matplotlib inline

# call function to warp image and pass original image
warped_image = warp(img)

f, (ax1, ax2) = plt.subplots(1, 2, figsize=(20,10))

ax1.set_title = "original image"
ax1.imshow(img)

ax2.set_title = "corrected image"
ax2.imshow(warped_image)

```
